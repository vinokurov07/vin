#Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
display(ratings_movies['title'].index)
```

```python
# Создайте в таблице новый признак year_release, который соответствует году выпуска фильма.
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
display(ratings_movies['year_release'].isna().sum())

```

```python
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?

mask1 = ratings_movies['year_release']==1999
display(ratings_movies[mask1].groupby(by='title')['rating'].mean().sort_values())
```

```python
# 8.3 Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
mask = ratings_movies['year_release']==2010
display(ratings_movies[mask].groupby(by='genres', as_index=False)['rating'].mean().sort_values(by='rating'))
display(ratings_movies[mask].groupby(by='genres')['rating'].mean().sort_values().reset_index())
```

```python
# 8.6 Найдите сочетание жанров (genres) за 2018 год, которое имеет наибольший средний рейтинг (среднее по столбцу rating),
# и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10.
mask = ratings_movies['year_release'] == 2018

ratings_movies[mask].groupby(by='genres')['rating'].agg(['mean','count']).sort_values(by=['mean','count'],ascending=[False,True])
df = ratings_movies[mask].groupby(by='genres')['rating'].agg(['mean','count']).sort_values(by=['mean','count'],ascending=[False,True])
display(df[df['count']>10])
```
