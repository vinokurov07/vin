### Чтоб из временного ряда вытащить например месяц-год, то применяют dt.to_period('M') (М тут - месяц). Получаем формат типа period, по нему график не постоить. dt.to_period('M'), действительно, даёт тип Period, который нельзя напрямую использовать в графиках. Можно либо преобразовать его в строку: solar_data['year-month'] = solar_data['date'].dt.to_period('M').astype(str)
или в datetime: solar_data['year-month'] = solar_data['date'].dt.to_period('M').dt.to_timestamp()

```python
solar_data['date'] = pd.to_datetime(solar_data['date'])
solar_data['year-month'] = solar_data['date'].dt.to_period('M')
solar_data['year-month'] = solar_data['date'].dt.to_period('M').astype(str)
solar_data['year-month'] = solar_data['date'].dt.to_period('M').dt.to_timestamp()
```
