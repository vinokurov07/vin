```python
import matplotlib.pyplot as plt
import numpy as np

# Настройка данных
x = np.linspace(0, 10, 100)

current_trend =  x*0.5  # Пример текущего тренда

# Разделение на исторические данные и прогноз
split_point = 5
x_past = x[x <= split_point]
x_future = x[x >= split_point]

# Сценарии прогноза
trend_continuation = 0.5 * x_future
upward_trend = 0.5 * split_point + 0.8 * (x_future - split_point)
downward_trend = 0.5 * split_point - 0.3 * (x_future - split_point)

# Создание графика
plt.figure(figsize=(10, 6))

# Исторические данные (сплошная линия)
plt.plot(x_past, current_trend[x <= split_point], 'b-', label='Historical trend')

# Прогнозные сценарии (пунктирные линии)
plt.plot(x_future, trend_continuation, 'b--', label='Сontinuation of the trend')
plt.plot(x_future, upward_trend, 'g--', label='Ascending forecast')
plt.plot(x_future, downward_trend, 'r--', label='Downward forecast')

# Вертикальная линия разделения
plt.axvline(split_point, color='gray', linestyle=':')

# Настройка отображения
plt.title('Forecast of consumption taking into account uncertainty')
plt.xlabel('Time')
plt.ylabel('Сonsumption, K m3/h')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
