```python
import matplotlib.pyplot as plt
import numpy as np

#  
x = np.linspace(0, 10, 100)

current_trend =  x*0.5  #   

#      
split_point = 5
x_past = x[x <= split_point]
x_future = x[x >= split_point]

#  
trend_continuation = 0.5 * x_future
upward_trend = 0.5 * split_point + 0.8 * (x_future - split_point)
downward_trend = 0.5 * split_point - 0.3 * (x_future - split_point)

#  
plt.figure(figsize=(10, 6))

#   ( )
plt.plot(x_past, current_trend[x <= split_point], 'b-', label='Historical trend')

#   ( )
plt.plot(x_future, trend_continuation, 'b--', label='ontinuation of the trend')
plt.plot(x_future, upward_trend, 'g--', label='Ascending forecast')
plt.plot(x_future, downward_trend, 'r--', label='Downward forecast')

#   
plt.axvline(split_point, color='gray', linestyle=':')

#  
plt.title('Forecast of consumption taking into account uncertainty')
plt.xlabel('Time')
plt.ylabel('onsumption, K m3/h')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
