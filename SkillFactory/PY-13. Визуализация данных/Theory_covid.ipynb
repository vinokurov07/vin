```python
import pandas as pd
covid_data = pd.read_csv('data/covid_data.csv', sep=',')
vaccinations_data = pd.read_csv('data/country_vaccinations.csv')
vaccinations_data = vaccinations_data[
    ['country', 'date', 'total_vaccinations', 
     'people_vaccinated', 'people_vaccinated_per_hundred',
     'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred',
     'daily_vaccinations', 'vaccines']]
```

#Предобработка данных

```python
covid_data = covid_data.groupby(by=['date','country'], as_index=False)[['confirmed', 'deaths', 'recovered']].sum()

covid_data['date'] = pd.to_datetime(covid_data['date'])
covid_data['active'] = covid_data['confirmed'] - covid_data['deaths'] - covid_data['recovered']

#Создадим признак ежедневного прироста числа заболевших, умерших и выздоровевших людей.
# Для этого отсортируем данные по названиям стран, а затем по датам. После этого произведём группировку по странам и рассчитаем 
# разницу между «вчера и сегодня» с помощью метода diff():
covid_data = covid_data.sort_values(by=['country', 'date'])
covid_data['daily_confirmed'] = covid_data.groupby('country')['confirmed'].diff()
covid_data['daily_deaths'] = covid_data.groupby('country')['deaths'].diff()
covid_data['daily_recovered'] = covid_data.groupby('country')['recovered'].diff()

vaccinations_data['date'] = pd.to_datetime(vaccinations_data['date'])

display(covid_data)
display(vaccinations_data)
#covid_vaccination = covid_data.merge(vaccinations_data, how='left', on='country')
covid_vaccination = covid_data.join(vaccinations_data.set_index(['country','date']), on=['country','date'], how='left', lsuffix='left')
display(covid_vaccination)
```

```python
covid_df = covid_data.merge(vaccinations_data, how='left', on=['date','country'])
print(covid_df.shape)
covid_df.to_csv('data\covid_df.csv')
```

```python
display(covid_df.head(1))
covid_df['death_rate'] = (covid_df['deaths'] / covid_df['confirmed'])*100
covid_df['recover_rate'] = (covid_df['recovered'] / covid_df['confirmed'])*100
covid_df.to_csv('data\covid_df.csv')
```
